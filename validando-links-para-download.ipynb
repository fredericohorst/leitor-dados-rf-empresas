# Checando por arquivos no Servidor Aberto da RF

Caso você queira fazer diversos downloads ao mesmo tempo - em vez de sequencial para cada arquivo que estão no servidor da Receita - esta solução ajuda a monitorar o endereço de download dos arquivos.

Endereço estilo FTP: http://200.152.38.155/

## Construindo um DataFrame contendo dados de todos os arquivos listados no Servidor da RF:

````python
import pandas
import numpy
import requests
from bs4 import BeautifulSoup

path = "http://200.152.38.155/CNPJ/"
df_files_rf = pandas.read_html(path)[0]
df_files_rf = df_files_rf.drop(0)
df_files_rf = df_files_rf.drop(1)
df_files_rf = df_files_rf.drop(columns='Unnamed: 0')
df_files_rf = df_files_rf.reset_index()
df_files_rf = df_files_rf.drop(columns='index')
df_files_rf.columns = ['Abrev_Name','LastModified','Size','Description']

response = requests.get(path)
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table')

links = []
for tr in table.findAll("tr"):
    trs = tr.findAll("td")
    for each in trs:
        try:
            link = each.find('a')['href']
            links.append(link)
        except:
            pass
links = links[1:]
diff = len(df_files_rf) - len(links)
if diff !=0:
    while len(df_files_rf) - len(links) > 0:
        links.append('NaN')

df_files_rf['CompleteName'] = links
````

## Construindo arquivos contendo no máximo 5 links de download:

````python

list_of_links = [path+name for name in df_files_rf.CompleteName]
list_of_links = list_of_links[:-1]

file = 'download_list_'
n=5
list = [list_of_links[i:i+n] for i in range(0, len(list_of_links), n)]
df = {}
df['index_no'] = numpy.array(range(len(list)))
df['lists'] = list
df = pandas.DataFrame(df)
# df.head(20)

for index, row in df.iterrows():
    i = row.index_no
    list = row.lists
    data = "\n".join(list)
    filename = file + str(i+1) + '.txt'
    with open(filename, 'w') as f:
        f.write(data)
    print(filename, " saved successfully! ")
````

## Pronto! Como usar os arquivos gerados:

Abra quantas janelas de um terminal bash desejar. Em cada uma delas, use: ```wget -i download_list_x.txt -P ~/caminho/completo/para/destino```, sendo `x` o número do arquivo.

Isso fará o terminal baixar todos os links dentro do arquivo mencionado.
